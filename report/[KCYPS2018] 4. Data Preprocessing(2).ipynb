## **1) 척도 방향 변경하기**

```python
def invert_scale_data(data, vars):
    for var in vars:
        data[var] = 5 - data[var]

    return data
```

```python
new_info = pd.read_excel('프로젝트_Codebook.xlsx')
new_info
```

```python
data_temp = data.copy()
vars = new_info.loc[new_info['방향']==-1,'변수명'].tolist()
new_data = invert_scale_data(data_temp, vars)
```

```python
# 검토
pd.concat([new_data[vars[0]].value_counts(),data[vars[0]].value_counts()],axis=1).sort_index()
```

## **2) 점수화하기(평균, 차이계산)**

```python
new_cat_dict = {'ID':['ID'],'ID_w':['ID_w']}
for i in range(new_info.shape[0]):
    name, code = new_info.iloc[i,:][['사용변수명','변수명']]

    if name in new_cat_dict.keys():
        new_cat_dict[name].append(code)
    else:
        new_cat_dict[name] = [code]
print(new_cat_dict)
```

```python
final_data = pd.DataFrame()
var_int_type = {}
for name, vars in new_cat_dict.items():
    if len(vars) == 1:
        final_data[name] = new_data[vars]
        if name not in ['ID','ID_w']:
            var_int_type.update({name:'int'})
    elif name in ['학업성취만족도_차이','학교생활만족도_차이']:
        final_data[name] = new_data[vars[0]]-new_data[vars[1]]
        var_int_type.update({name:'int'})
    else:
        final_data[name] = data[vars].mean(axis=1)
final_data.columns
```
