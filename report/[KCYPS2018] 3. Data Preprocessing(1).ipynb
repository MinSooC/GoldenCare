# **2. 결측치 처리**

```python
def check_NA(data, axis=1):
    nrow = data.shape[axis]
    count_NA = data.isna().sum(axis=axis)
    ratio_NA = count_NA / nrow

    report_NA = pd.concat([count_NA, ratio_NA], axis=1)
    report_NA.index.name = 'variables'
    report_NA.columns = ['count', 'ratio']

    if report_NA['count'].sum() == 0:
        return '결측치가 없습니다.'
    else:
        # print(f"결측치가 총 {report_NA['count'].sum()} 개 존재합니다.")
        report_NA = report_NA.loc[report_NA['count'] > 0,:]

    return report_NA.sort_values(by='count', ascending=False)
```

## **1) 열 데이터의 결측치 파악하기**

```python
# 열별 결측치 개수에 따른 카운트
bycol_check_na = check_NA(data, axis=0)
bycol_check_na['count'].value_counts().sort_index(ascending=False)
```

## **2) 행 데이터의 결측치 파악하기**

```python
# 행별 결측치 개수에 따른 카운트
byrow_check_na = check_NA(data)
byrow_check_na['count'].value_counts().sort_index(ascending=False)
```

### 과반 수 이상의 결측인 행(ID)는 '무응답자'라고 정의하고 이를 삭제하였다.

```python
# 무응답자 ID 추출
index = byrow_check_na.loc[byrow_check_na['count'] > 168,:].index
rem_id_lst = data.iloc[index]['ID_w'].tolist()
print(rem_id_lst)
print(f'무응답자 = {len(rem_id_lst)}')
```

```python
# ID를 가지고 있는 행 추출
remove_idx = []
for ind, x in enumerate(data['ID_w']):
    if x in rem_id_lst:
        remove_idx.append(ind)
len(remove_idx)
```

```python
# 무응답 제거
print(f'Before: {data.shape}')
data = data.drop(index=remove_idx)
print(f'After : {data.shape}')
```

```python
# Reset index
data = data.reset_index(drop=True)
data.shape
```
