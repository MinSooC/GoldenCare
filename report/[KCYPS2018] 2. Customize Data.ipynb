# **3. 변수 선정 기준 정하기(조사항목을 기준으로 함)**

* 공통(배경변인) : 개인의 기본정보를 포함하고자 함.<br>
$\longrightarrow$ id, 나이, 거주지, 월평균가구소득, 월평균 교육비
* 아동
    1. 개인 발달(초등,중등) : 아동 심리에 영향을 끼칠 수 있는 변수를 포함하고자 함.<br>
$\longrightarrow$ 부모님 대화시간, 학습시간, 여가시간, 학업 만족도<br>
$\longrightarrow$ 만족도, 행복, 자아인식, 끈기(공통)<br>
$\longrightarrow$ 정서문제, 협동의식
    2. 발달 환경(초등,중등) : 아동 심리에 영향을 끼칠 수 있는 변수를 포함하고자 함.<br>
$\longrightarrow$ 스마트폰(공통)<br>
$\longrightarrow$ 학교생활, 양육태도<br>

* 보호자
    1. 보호자관점 : 아동과 공통된 질문으로 선택하여 비교하고자 함.<br>
$\longrightarrow$ 정서문제, 협동의식
    2. 보호자개인 : 아동의 정서와 부모의 정서 관계를 비교하고자 함.<br>
$\longrightarrow$ 만족도, 행복, 자아인식, 끈기(공통)<br>
$\longrightarrow$ 스마트폰(공통)<br>


# **4. 선정 변수 추출 및 변환하기**

```python
def get_search_var(var_dict, cat_dict, subject, var_hangul_lst):
    temp = [(var_dict[subject][x],x) for x in cat_dict]

    return [v for k,v in temp if k in var_hangul_lst]
```

```python
def make_hangul_var_name(name_lst, var_lst):
    var_name = []
    for i, temp in enumerate(var_lst):
        if type(temp) == str:
            var_name.append((temp,name_lst[i]))
        else:
            var_name.extend([(x,name_lst[i]+str(j)) for j,x in enumerate(temp, 1)])

    return var_name
```

```python
def get_under_var(var_dict, cat_dict, subject):
    temp = [(var_dict[subject][x],x) for x in cat_dict]

    return [v for k,v in temp]
```
