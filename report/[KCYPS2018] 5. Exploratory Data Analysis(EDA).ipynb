## **2) 결측치가 있나요?**

```python
# 결측치 확인
def check_NA(data, axis=1):
    nrow = data.shape[axis]
    count_NA = data.isna().sum(axis=axis)
    ratio_NA = count_NA / nrow

    report_NA = pd.concat([count_NA, ratio_NA], axis=1)
    report_NA.index.name = 'variables'
    report_NA.columns = ['count', 'ratio']

    if report_NA['count'].sum() == 0:
        return '결측치가 없습니다.'
    else:
        # print(f"결측치가 총 {report_NA['count'].sum()} 개 존재합니다.")
        report_NA = report_NA.loc[report_NA['count'] > 0,:]

    return report_NA.sort_values(by='count', ascending=False)
```

```python
check_NA(data)
```

```python
from scipy import stats
data_m = data.loc[data['성별']==1,:]
data_f = data.loc[data['성별']==2,:]

stats.ttest_ind(data_m['아동_정서문제'], data_f['아동_정서문제'], equal_var=True)
```
